```python
import numpy as np

from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error
```

```python


from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y_trns,test_size=0.2, random_state=1)
```

```python
transformer= ColumnTransformer(transformers=   [
    ("tnf1_OHE", OneHotEncoder(handle_unknown='ignore'),['Company', 'TypeName','OpSys','Memory_type','processor_name','Gpu_name']),
    ('tnf2_MMS', MinMaxScaler(),['x_res', 'y_res','Ram','Total_Memory_GB', 'Weight','Inches','processor_speed_Gigahrtz'])
     ],remainder="passthrough"
    
                      )
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.ensemble import RandomForestRegressor
RR=RandomForestRegressor(max_features= 0.90,

 max_samples= 0.80,
 n_estimators= 300,          #The number of trees in the forest.

min_samples_split= 10
                       )
```

MAKING PIPELINE INCORPORATING COLUM TRANSFORMER AND THEN THE ML ALGORITHM 

```python
model = Pipeline(steps=[('transformer', transformer), ('regressor', RR)])
```

```python
model.fit(x_train, y_train )
```

```python
y_pred= model.predict(x_test)
```

```python
#checking parameters 
#mean square error  and R value 

mse_RR=mean_squared_error(y_test,y_pred)
print('mse_RR', mse_RR)
maeRR= mean_absolute_error(y_test,y_pred)
print('maeRR',maeRR)
np.sqrt(mse_RR)
print('MSRE',np.sqrt(mse_RR))
print('r2score',r2_score(y_test, y_pred))

#with Linear regression the mse was 301904185.58954406
#nestimator inc the r2 score a little

```

```python

def predict_laptop_price(Company, TypeName, Inches, Ram, OpSys, Weight, Total_Memory_GB, Memory_type, x_res, y_res, Touchscreen, IPS_Panel, HD, processor_name, processor_speed_Gigahrtz, Gpu_name):
    inputs = pd.DataFrame([[Company, TypeName, Inches, Ram, OpSys, Weight,
                            Total_Memory_GB, Memory_type, x_res, y_res, Touchscreen,
                            IPS_Panel, HD, processor_name, processor_speed_Gigahrtz, Gpu_name]],
                          columns=['Company', 'TypeName', 'Inches', 'Ram', 'OpSys', 'Weight',
                                   'Total_Memory_GB', 'Memory_type', 'x_res', 'y_res', 'Touchscreen',
                                   'IPS_Panel', 'HD', 'processor_name', 'processor_speed_Gigahrtz', 'Gpu_name'])

    price = np.expm1(model.predict(inputs)[0])
    return price
    
```

```python
company= gr.Dropdown(label='company', choices = ['Lenovo', 'Dell', 'HP', 'Asus', 'Acer','others'])
TypeName = gr.Dropdown(label='Type_name', choices = ["Notebook", "Gaming", "Ultrabook", "2 in 1 Convertible", "Workstation", "Netbook"])
Inches= gr.Number(label='Inches of laptop')
Ram= gr.Number(label='Enter GBs of RAM in multiples of 4 and 8')
OpSys= gr.Dropdown(label= 'OpSys',choices = ["Windows","Linux", "Mac", "Chrome OS", "No OS", "others"])
Weight= gr.Number(label='weight in Kgs')
Total_Memory_GB= gr.Number(label='Memory in GBs')
Memory_type= gr.Dropdown(label='Memory_type',choices = ["SSD memory", "HDD memory", "Flash Storage", "Hybrid memory"])
x_res= gr.Number(label='enter first pixel resolution (4 figure)')
y_res= gr.Number(label='enter first pixel resolution (4 figure)')
Touchscreen= gr.Number(label='Touchscreen (0= NO, 1= yes)')
IPS_Panel= gr.Number(label='IPS_Panel (0= NO, 1= yes)')
HD=gr.Number(label='HD (0= NO, 1= yes)')
processor_name= gr.Dropdown(label='processor_name', choices= ['Intel Core i7', 'Intel Core i5', 'Other Intel Processor', 'Intel Core i3', 'AMD Processors', 'Other Processors'])
processor_speed_Gigahrtz= gr.Number(label='speed in float descimal (2 figures)')
Gpu_name= gr.Dropdown(label='Gpu_name',choices = [ "Intel HD Graphics", "Nvidia GeForce GPU", "AMD Radeon GPU","Intel UHD Graphics"," Nvidia Quadro GPU","Other intel GPU","Other AMD GPU", "Other Nvidia GPU", "Other GPU"])
```

```python
demo=gr.Interface(fn=predict_laptop_price,inputs=[company, TypeName, Inches, Ram, OpSys, Weight,
                            Total_Memory_GB, Memory_type, x_res, y_res, Touchscreen,
                            IPS_Panel, HD, processor_name, processor_speed_Gigahrtz, Gpu_name],outputs=gr.Textbox(label="Predicted Laptop Price"))
```

```python
demo.launch(share=True)
```
